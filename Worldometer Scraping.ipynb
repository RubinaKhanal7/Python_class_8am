```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.worldometers.info/coronavirus', 'html.parser')
soup = requests.get(url)
soup
```

```python
code = BeautifulSoup(soup.text,"lxml")
code = code.find("table", class_="table table-bordered table-hover main_table_countries")
tr_code = code.find_all("tr")
tr_code
```

```python
rows = []
for i in tr_code[1:]:
    tds = []
    for j in i:
        z = j.text.replace(" ","").replace('\n','')
        if z != "":
            tds.append(z)
    
    rows.append(tds)
rows
```

```python
header = tr_code[0].text.split('\n')[1:-1]
rows = [row[:len(header)] + [""] * (len(header) - len(row)) for row in rows]
```

```python
import pandas as pd

df = pd.DataFrame(rows,columns = header)
df
```

```python
df.to_csv('worldometer_data.csv')
```

```python
df['TotalCases'] = [float(i.replace(',','')) for i in df['TotalCases']]
```

```python
df1 = df
```

```python
df1 = df.sort_values(by=['TotalCases'], ascending = False).iloc[0:10]
df1
```

```python
df1
```

```python
import plotly.express as px

fig = px.bar(df1, x='Country,Other', y='TotalCases')
fig.show()
```

```python
df1['ActiveCases'] = [float(i.replace(',','')) for i in df1['ActiveCases']]
df1['Serious,Critical'] = [float(i.replace(',','')) for i in df1['Serious,Critical']]
```

```python
df1
```

```python
import plotly.graph_objects as go


fig = go.Figure(data=[
    go.Bar(name='Active Cases', x=df1['Country,Other'], y=df1['ActiveCases']),
    go.Bar(name='Serious or Critical', x=df1['Country,Other'], y=df1['Serious,Critical'])
])
fig.update_layout(barmode='group')
fig.show()
```
