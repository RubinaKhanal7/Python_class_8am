```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://nphl.gov.np/patient-portal', 'html.parser')
soup = requests.get(url)
soup
```

```python
code = BeautifulSoup(soup.text,"lxml")
code = code.find("table", class_="table table-striped table-hover")
tr_code = code.find_all("tr")
tr_code
```

```python
import pandas as pd

data = []
for i in tr_code:
    y = i.text.split('\n')
    try:
        y[3] = int(y[3])
        data.append(y[1:-1])
    except:
        data.append(y[1:-1])
        continue
df = pd.DataFrame(data[1:], columns = data[0])
df
```
